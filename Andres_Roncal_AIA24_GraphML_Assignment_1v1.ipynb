```python
import networkx as nx
import matplotlib.pyplot as plt
import random
#Check nx version
nx.__version__

G = nx.DiGraph() #empty graph Graph DiGraph
plt.figure(figsize=(15, 10))


me = ['Andres']
workFriends = ['John', 'Jane', 'Michael', 'Emily', 'Christopher']
uniFriends = ['Jessica', 'Matthew', 'Emma', 'David', 'Olivia']
mtFriends = ['Andrew', 'Ava', 'Daniel', 'Sophia', 'James']
lifeFriends = ['Joseph', 'Charlotte', 'William', 'Amelia', 'Benjamin']

V1 = me + workFriends + uniFriends + me + mtFriends + me + lifeFriends
V2 = uniFriends + me + workFriends + me + lifeFriends + me + mtFriends
```

```python
friendPairs = {friend1: friend2 for friend1, friend2 in zip(V1, V2)}
namePairs = list(friendPairs.items())
print(friendPairs)
print(namePairs)
```

```python
# assigning colors to each group as a dictionary then make into tuple and add random weight
colorsd = {}
for friend in V:
    if friend in me:
        colorsd[friend] = "#8B0000"
    elif friend in workFriends:
        colorsd[friend] = "#9A32CD"
    elif friend in uniFriends:
        colorsd[friend] = "#AB82FF"
    elif friend in mtFriends:
        colorsd[friend] = "#BA55D3"
    elif friend in lifeFriends:
        colorsd[friend] = "#DA70D6"
colors = list(colorsd.items())
E = [(name, color, random.randint(1, 50)) for name, color in colors]
colors = [item[1] for item in E]
print(colors)
```
